```python
np.count_nonzero((df['Age']>0) & (df['Age']<20))
np.count_nonzero(df['Age']<20)
np.sum((df['Age']>0) & (df['Age']<20))
```

```python
b_y=np.count_nonzero((df['Age']>0) & (df['Age']<=20))
b_a=np.count_nonzero((df['Age']>20) & (df['Age']<=40))
b_m=np.count_nonzero((df['Age']>40) & (df['Age']<=60))
b_o=np.count_nonzero(df['Age']>60)
print(b_y,b_a,b_m,b_o)
```

```python
labels=['Young','Adult','MiddleAge','OldAge']
plt.bar(labels,(b_y,b_a,b_m,b_o),color=['r','g','b','m'])
plt.show()
```
