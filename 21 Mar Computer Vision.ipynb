```python
img=cv2.imread("/content/sudoku.png",0)
cv2_imshow(img)
```

```python
img=cv2.resize(img,(280,250))
cv2_imshow(img)
```

```python
sobel_x=cv2.Sobel(img,cv2.CV_8U,dx=1,dy=0,ksize=-1)
sobel_y=cv2.Sobel(img,cv2.CV_8U,dx=0,dy=1,ksize=-1)
sobel_f=cv2.bitwise_or(sobel_x,sobel_y)
laplacian=cv2.Laplacian(img,cv2.CV_8U)
canny=cv2.Canny(img,250,150)
cv2_imshow(sobel_x)
cv2_imshow(sobel_y)
cv2_imshow(sobel_f)
cv2_imshow(laplacian)
cv2_imshow(canny)
```

```python
#Drawing lines
a=np.zeros([300,300,3],dtype="uint8")
cv2.line(a,(0,50),(150,50),(255,255,255),5)
cv2.line(a,(150,50),(150,150),(255,0,255),5)
cv2.line(a,(150,150),(0,50),(0,255,255),5)
cv2.line(a,(0,50),(0,150),(255,255,0),5)
cv2.line(a,(0,150),(150,150),(0,0,255),5)
cv2_imshow(a)
```
