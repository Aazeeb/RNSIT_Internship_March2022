```python
#Dataframe with values from a dictionary
#handling missing values
```

```python
d={'Cricket':[1,3,np.nan,4,6,7,2,np.nan],
   'Baseball':[5,np.nan,np.nan,5,7,4,7,3],
   'Tennis':[1,2,3,4,5,6,7,8]}
```

```python
df=pd.DataFrame(d)
df
```

```python
#Handling of missing data
#(1) Drop the rows or colums containing the missing values(dropna)
#(2) Fill the missing values with 0 (fillna)
#(3) Fill the missing values with mean or median or mode (fillna)
```

```python
df2=df.dropna(axis=0)  #dropping the null values along the rows
df2
```

```python
df.dropna(axis=1) #dropping the null values along the columns
```

```python
df.fillna(value=0)
```

```python
df['Cricket']=df['Cricket'].fillna(value=df['Cricket'].median())
```

```python
df
```

```python
df['Baseball']=df['Baseball'].fillna(value=df['Baseball'].median())
df
```

```python
df=df.set_index([['Alpha','Bravo','Charlie','Delta','Echo','Foxtrot','Golf','Hotel']])
df
```

```python
df=df.reset_index()
#df.reset_index(inplace=True)
df
```
