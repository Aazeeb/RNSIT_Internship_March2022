```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/drive/MyDrive/RNSIT March17/Salary_Data.csv")
df
```

```python
x=df.iloc[:,0:1].values
y=df.iloc[:,1].values
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression as LR
#from sklearn import linear_model.LinearRegression as LR
#import sklearn.linear_model.LinearRegression
model=LR()
```

```python
model.fit(x,y)
```

```python
model.predict([[9.6]])
```

```python
model.coef_  #slope
```

```python
model.intercept_  #c
```

```python
#Visualization
```

```python
y_pred=model.predict(x)
y_pred
```

```python
y
```

```python
plt.figure(figsize=(9,8))
plt.scatter(x,y,color="r",label="Actual Values",marker="*",alpha=0.6)
plt.plot(x,y_pred,color="k",label="Predicted Values",marker=".")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary Prediction based on years of experience")
plt.legend()
plt.xticks(np.arange(1,10.6,0.5),rotation=60)
plt.yticks(np.arange(35000,125000,5000))
plt.grid()
plt.show()
```

```python
#Evaluation
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y,y_pred)
```

```python
r2_score(y,y_pred)
```
