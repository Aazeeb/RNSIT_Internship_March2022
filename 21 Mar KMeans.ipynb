```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/drive/MyDrive/RNSIT March17/Mall_Customers.csv")
df.head()
```

```python
df.shape
```

```python
x=df.iloc[:,3:5].values
x
```

```python
#below block of code is to determine the best value of k(no. of clusters)
#we can determine k by using elbow method
#max value of K shuold not exceed the square root of total number of rows
```

```python
k=range(2,15)
from sklearn.cluster import KMeans
SSE=[]
for i in k:
  model_demo=KMeans(n_clusters=i)
  model_demo.fit(x)
  sse=model_demo.inertia_
  SSE.append(sse)

plt.plot(k,SSE)
plt.show()
```

```python
#as per the above graph, th elbow is steepest at 5
#hence, we will take the no. of clusters as 5
```

```python
k=5
from sklearn.cluster import KMeans
model=KMeans(n_clusters=k,random_state=0)
model.fit(x)
```

```python
y=model.predict(x)
y
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x[y==0,0],x[y==0,1],label="Cluster 0")
plt.scatter(x[y==1,0],x[y==1,1],label="Cluster 1")
plt.scatter(x[y==2,0],x[y==2,1],label="Cluster 2")
plt.scatter(x[y==3,0],x[y==3,1],label="Cluster 3")
plt.scatter(x[y==4,0],x[y==4,1],label="Cluster 4")
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],
            label="Centroids",s=200,c="Yellow",alpha=0.4)
plt.legend()
plt.xlabel("Annual Income(k$)")
plt.ylabel("Spending Score(1-100)")
plt.title("Market Basket Analysis")
plt.show()
```
