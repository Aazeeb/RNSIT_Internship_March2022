```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/drive/MyDrive/RNSIT March17/knn.csv")
df
```

```python
x=df.iloc[:,0:2].values
y=df.iloc[:,2].values
```

```python
y
```

```python
k=5 #square root of total number of records/rows
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNC
```

```python
model=KNC(n_neighbors=k,metric="euclidean")
```

```python
model.fit(x,y)
```

```python
pred=[[55.2,176]]
```

```python
model.predict(pred)
```
