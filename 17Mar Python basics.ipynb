```python
#write a code to determine the grade of students scoring betwn 0 and 100
#90 and above is excellent
#70 and 90 is good
#50 and 70 is average
#below 50 is needs improvement
```

```python
marks= int(input("Enter the marks scored in whole numbers(0-100): "))
if(marks>100):
  print("Invalid marks, please rerun the cell and try again")
elif(marks>=90):
  print("Excellent")
elif(marks>=70):
  print("Good")
elif(marks>=50):
  print("average")
elif(marks>0):
  print("Needs improvement")
else:
  print("Invalid marks, please rerun the cell and try again")
```

```python
#Use nested if, else conditions to find the greatest of 3 numbers
#use input statements to get the numbers
```

```python
num1=int(input("Enter the first number: "))
num2=int(input("Enter the second number: "))
num3=int(input("Enter the third number: "))
if(num1<num2):
  if(num2<num3):
    print("Number 3 is the greatest")
  else:
    print("Number 2 is the greatest")
else:
  if(num1>num3):
    print("Number 1 is the greatest")
  else:
    print("Number 3 is the greatest")
```
