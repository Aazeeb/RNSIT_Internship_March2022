```python
#Regression vs Forecasting
#Regression is prediction of a value with respect to another continous value
#Forecasting is prediction of a value with respect to time or date
```

```python
#Prophet developed by FB
#The input given to fbprophet should be a dataframe
#The dataframe should have 2 columns, named ds and y
#The ds column(date) should be of the datatype "datetime"
#The y column should be a numeric datatype
```

```python
df.head()
```

```python
df.rename(columns={'Modal Price (Rs/q)':'y','Date':'ds'},inplace=True)
df.head()
```

```python
df_new=df[['ds','y']]
df_new.head()
```

```python
df_new.info()
```

```python
#columns are not in required formats
#convert ds into datetime format
#convert y into int format
```

```python
df_new['y']=df_new['y'].astype(int)
df_new['ds']=pd.to_datetime(df_new['ds'])
```

```python
df_new.info()
```

```python
df_new.head()
```

```python
df_new.isnull().sum()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.scatter(df_new['ds'],df_new['y'])
plt.show()
```

```python
import numpy as np
df_new['y']=np.where(df_new['y']>3000,np.nan,df_new['y'])
df_new
```

```python
df_new.isnull().sum()
```

```python
df_new=df_new.dropna()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.scatter(df_new['ds'],df_new['y'])
plt.show()
```

```python
import fbprophet
op=fbprophet.Prophet(changepoint_prior_scale=1)
```

```python
op.fit(df_new)
```

```python
forecast=op.make_future_dataframe(periods=365,freq="D")
forecast=op.predict(forecast)
```
