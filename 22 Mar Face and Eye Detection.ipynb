```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
!wget https://www.quickanddirtytips.com/sites/default/files/images/2332/people-persons-peoples.jpg -O p1.jpg
```

```python
img=cv2.imread("/content/p1.jpg")
cv2_imshow(img)
```

```python
img.shape
```

```python
img=cv2.resize(img,(500,350))
cv2_imshow(img)
```

```python
!wget https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml
```

```python
model=cv2.CascadeClassifier("/content/haarcascade_frontalface_default.xml")
face=model.detectMultiScale(img,1.1,2)
for (x,y,w,h) in face:
  cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
cv2_imshow(img)
```
