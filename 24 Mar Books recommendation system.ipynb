```python
import numpy as np
import pandas as pd
books=pd.read_csv("/content/drive/MyDrive/RNSIT March17/BX-Books.csv"
,sep=";",encoding="latin-1",error_bad_lines=False)
users=pd.read_csv("/content/drive/MyDrive/RNSIT March17/BX-Users.csv"
,sep=";",encoding="latin-1",error_bad_lines=False)
ratings=pd.read_csv("/content/drive/MyDrive/RNSIT March17/BX-Book-Ratings.csv"
,sep=";",encoding="latin-1",error_bad_lines=False)
```

```python
books=books[['ISBN','Book-Title','Book-Author','Year-Of-Publication','Publisher']]
books.rename(columns={'Book-Title':'title','Book-Author':'author',
                     'Year-Of-Publication':'year','Publisher':'publisher'},inplace=True)
users.rename(columns={'User-ID':'user_id','Location':'location','Age':'age'},inplace=True)
ratings.rename(columns={'User-ID':'user_id','Book-Rating':'rating'},inplace=True)            
```

```python
#limit the users to ones who has rated atleast 200 books
#limit the books to ones which have received atleast 50 ratings
```

```python
ratings['user_id'].value_counts()
```

```python
x=ratings['user_id'].value_counts()>200
y=x[x].index
print(y.shape)
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings_with_books=ratings.merge(books,on='ISBN')
ratings_with_books.head(2)
```

```python
number_rating=ratings_with_books.groupby('title')['rating'].count().reset_index()
number_rating
```

```python
number_rating.rename(columns={'rating':'number_of_ratings'},inplace=True)
final_rating=ratings_with_books.merge(number_rating,on='title')
```

```python
final_rating.shape
```

```python
final_rating=final_rating[final_rating['number_of_ratings']>=50]
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating
```

```python
book_pivot=final_rating.pivot_table(index="title",columns="user_id",values="rating")
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm="brute")
model.fit(book_sparse)
```

```python
book_pivot.iloc[0,:].values.reshape(1,-1)
```

```python
dist, suggestions=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1))
```
