```python
e_bangalore = [100.6, 158.61, 305.54, 394.96, 724.79]
e_mysore =    [10.5,  25.21,  58.65,  119.27, 274.87]
year=         [1972,  1982,   1992,   2002,   2012  ]
```

```python
np.arange(1970,2014,4)
```

```python
plt.plot(year,e_bangalore)
plt.plot(year,e_mysore)
```

```python
plt.figure(figsize=(7,5))
plt.plot(year,e_bangalore,c="r",label="Bangalore",marker="*",linestyle="dashed")
plt.plot(year,e_mysore,c="k",label="Mysore",marker=".",linestyle="dashdot")
plt.xlabel("Years",fontsize=13)
plt.ylabel("Electricity consumption",fontsize=13)
plt.title("Electricity consumption trend analysis over the years",fontsize=14)
plt.legend(loc="upper left")
plt.xticks(year,rotation=45)
#plt.xticks(np.arange(1970,2014,4),rotation=45)
plt.grid()
plt.savefig("out.png")
plt.show()
```

```python
cars =  ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES' ]
data=   [   23,    17,   35,    29,      12,        41   ]
color=  ['Orange',  'Red',  'Green',  'Blue', 'Cyan', 'Magenta']
explode=[       0.2,      0,      0,      0,      0.4,      0]
```

```python
plt.figure(figsize=(7,6))
plt.pie(data,labels=cars,colors=color,counterclock=False,
        startangle=90,shadow=True, autopct="%1.1f%%",explode=explode,
        wedgeprops={"edgecolor":"black","linewidth":2})
plt.legend()
plt.show()
```
